```python
import pandas as pd
import glob
import torch

# dataframe for sharpe ratio and total return
sharpe_and_return_df = pd.read_csv(f'backtesting_result/sharpe_and_return.csv', encoding = 'utf-8', encoding_errors = 'ignore')

# dataframe for test loss
test_loss_df = pd.read_csv(f'test_loss/test_loss.csv', encoding = 'utf-8', encoding_errors = 'ignore')

# dataframe for 成交金額
# volume_df = pd.read_excel(f'月成交金額/月成交金額2023.xlsx')
# volume_df.rename(columns={'股票代號': 'stock_number'}, inplace=True)

# dataframe for 類股資訊
category_df = pd.read_excel(f'類股/產業分類.xlsx')
category_df.rename(columns={'股票代號': 'stock_number'}, inplace=True)
  
merged_df = pd.merge(sharpe_and_return_df, test_loss_df, on='stock_number', how='outer')
# merged_df = pd.merge(merged_df, volume_df, on='stock_number', how='outer')
merged_df = pd.merge(merged_df, category_df, on='stock_number', how='outer')
sorted_df = merged_df.sort_values(by=['sharpe_ratio'], ascending=[False])
# 移除那些 'Unnamed' 開頭的列
sorted_df = sorted_df.loc[:, ~sorted_df.columns.str.contains('^Unnamed')]
sorted_df = sorted_df.reset_index(drop = True)
sorted_df = sorted_df[sorted_df['total_return']>=2]
sorted_df = sorted_df[sorted_df['test_mse_loss']<=0.002]
sorted_df = sorted_df[sorted_df['2024產業名稱']=='電子–半導體']
sorted_df.to_csv(f'selected_stock/sorted_stock.csv')
'''
# select top ten stock each month by volume
for i in range(1,13):
    if i in range(1,10):
        selected_df = sorted_df[sorted_df[f'20230{i}成交金額(千)']>=30000]
    else:
        selected_df = sorted_df[sorted_df[f'2023{i}成交金額(千)']>=30000]
    top_10_stocks = selected_df.head(10)
    top_10_stocks.to_csv(f'selected_stock/final_selection/selected_stock{i}.csv')
'''   
top_5_stocks = sorted_df.head(5)
top_5_stocks.to_csv(f'selected_stock/selected_stock.csv')
top_5_stocks
```
